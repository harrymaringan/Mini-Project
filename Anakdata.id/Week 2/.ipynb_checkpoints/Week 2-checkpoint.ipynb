*Call Item Data*

```python
item =pd.read_excel(r"E:/anakdata.id/Week 2/Item.xlsx")
```

```python
item.head()
```

```python
item.isnull().sum(0)
```

```python
transaction_cust_item = pd.merge(item, transaction_cust, on=["ITEM_CODE"], how= "outer", validate ="one_to_many")
```

```python
transaction_cust_item.head(10)
```

```python
transaction_cust_item.isnull().sum()
```

```python
transaction_cust_item.set_index("CUSTOMER_ID",inplace=True)
```

```python
transaction_cust_item.tail(10)
```

```python
transaction_cust_item.info()
```

```python
df = transaction_cust_item.drop(transaction_cust_item.tail(20).index)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.head()
```

```python
df["TOTAL_REVENUE"] = df.QTY*df.SELLING_PRICE
```

```python
df.head()
```

```python
df["NET_PROFIT"] = (df.SELLING_PRICE - df.PRODUCTION_PRICE)*df.QTY
```

```python
df.head()
```

*Calculate Age*

```python
from datetime import datetime
df["BORN_DATE"] = pd.to_datetime(df["BORN_DATE"], errors="coerce")
now =pd.to_datetime('now')
now
```

```python
df["AGE"] = (now.year - df["BORN_DATE"].dt.year) - ((now.month - df["BORN_DATE"].dt.month) < 0 )
```

```python
df.head()
```

*Calculate Tenure*

```python
df["TENURE"] = (now.month - df["REGISTRATION_DATE"].dt.month)
```
