```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_excel(r"E:/anakdata.id/Week 2/Sales_Cleaned.xlsx")
```

```python
df.info()
```

```python
df["Quarter"] = df.TRANSACTION_DATE.dt.quarter
df["Week"] = df.TRANSACTION_DATE.dt.week
df["Weekday"] = df.TRANSACTION_DATE.dt.weekday
df["Day"] = df.TRANSACTION_DATE.dt.day
df["Dayofyear"] = df.TRANSACTION_DATE.dt.dayofyear
```

```python
grouped_features = ["TRANSACTION_DATE", "year", "Quarter","month", "Week", "Weekday", "Dayofyear", "Day",
                    "ITEM_CODE"]
```

```python
daily_data = pd.DataFrame(df.groupby(grouped_features).QTY.sum(),
                          columns=["QTY"])
daily_data["TOTAL_REVENUE"] = df.groupby(grouped_features).TOTAL_REVENUE.sum()
daily_data = daily_data.reset_index()
daily_data.head(15)
```

```python
daily_data.loc[:, ["QTY", "TOTAL_REVENUE"]].describe()
```

```python
low_quantity = daily_data.QTY.quantile(0.01)
high_quantity = daily_data.QTY.quantile(0.99)
print((low_quantity, high_quantity))
```

```python
low_revenue = daily_data.TOTAL_REVENUE.quantile(0.01)
high_revenue = daily_data.TOTAL_REVENUE.quantile(0.99)
print((low_revenue, high_revenue))
```

```python
samples = daily_data.shape[0]
```

```python
daily_data = daily_data.loc[
    (daily_data.QTY >= low_quantity) & (daily_data.QTY <= high_quantity)]
daily_data = daily_data.loc[
    (daily_data.TOTAL_REVENUE >= low_revenue) & (daily_data.TOTAL_REVENUE <= high_revenue)]
```

```python
samples - daily_data.shape[0]
```

```python
fig, ax = plt.subplots(1,2,figsize=(20,5))
sns.distplot(daily_data.QTY.values, kde=True, ax=ax[0], color="Orange", bins=30);
sns.distplot(np.log(daily_data.QTY.values), kde=True, ax=ax[1], color="Orange", bins=30);
ax[0].set_xlabel("Number of daily product sales");
ax[0].set_ylabel("Frequency");
ax[0].set_title("How many products are sold per day?");
```
